```python
df.rename(columns = {'ï»¿Average Taxable Income $':'Average Taxable Income'}, inplace = True)
df.head()
```

```python
df['Average Taxable Income'] = df['Average Taxable Income'].str.replace('$', '')
df['Average Taxable Income'] = df['Average Taxable Income'].str.replace(',', '')
df.head()
```

```python
df['Average Taxable Income']=df['Average Taxable Income'].astype('int64')
df.dtypes
```

```python
df["Female Income"] = df.eval("Gender == 'Female'")
df["Female Income"] = df[df['Average Taxable Income'].notnull() & (df['Gender'] == "Female")]
df.head()
```

```python
df["Male Income"] = df.eval("Gender == 'Male'")
df.tail()
```

```python
df.rename(columns = {'Male Income':'Male_Income',
                     'Average Taxable Income':'Average_Taxable_Income',
                     'Female Income':'Female_Income'
                    }, inplace = True)
df.head()
```

```python
male_occ = df[df.Gender == "Male"].Occupation
female_occ = df[df.Gender == "Female"].Occupation
gender_occ = list(set(male_occ).intersection(set(female_occ)))
df["Male_Income"] = df[(df.Occupation.isin(gender_occ)) & 
                       df['Average_Taxable_Income'].notnull() & 
                       (df['Gender'] == "Male")]
female_sal = df[df.Gender == "Female"].Average_Taxable_Income
male_sal = df[df.Gender == "Male"].Average_Taxable_Income
#(df.Male_Income.fillna(df.Average_Taxable_Income , inplace=True))
df.Male_Income.fillna((df.Average_Taxable_Income.isin(male_sal&male_occ)) , inplace=True)
df.head()
#df[df['Average Taxable Income'].notnull() & (df['Gender'] == "Male") & (df.Occupation.isin(gender_ambiguous_names))]
#df["Male Income"] = df[df['Average Taxable Income'].notnull() & (df['Gender'] == "Male") & (df['Occupation'](df['Gender']=="Female") == df['Occupation'](df['Gender']=="Male"))]
#df.loc[df['Gender'] == "Female" & df['Occupation'].isin(Female)]
#df.loc[df.Gender == 'Female' & , 'FirstName'] = "Matt"
#df.loc[df.ID == 103, 'LastName'] = "Jones"
```
